```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files

# Upload file
uploaded = files.upload()

# Load the dataset
df = pd.read_excel('who_data.xlsx')

# Extract the data for the bar chart
years = df.columns[4:]  # The years are in columns from index 4 onwards
female_population_percentage = df.iloc[0, 4:].values  # Values of the female population for each year

# Print the extracted data to check
print("Extracted Years:", years)
print("Female Population Percentages:", female_population_percentage)

# Configuration for plots using a dictionary (JSON-like)
plot_config = {
    "figure_size": (12, 10),
    "bar_color": 'red',
    "hist_color": 'yellow',
    "hist_bins": 10,
    "hist_edgecolor": 'black'
}

# Create a figure with two subplots: one for bar chart, one for histogram
plt.figure(figsize=plot_config["figure_size"])

# First subplot: Bar chart
plt.subplot(2, 1, 1)  # 2 rows, 1 column, 1st subplot
plt.bar(years, female_population_percentage, color=plot_config["bar_color"])
plt.title('Female Population Ages 20-24 in India (Bar Chart)')
plt.xlabel('Year')
plt.ylabel('Female Population (%)')
plt.xticks(rotation=90)

# Second subplot: Histogram
plt.subplot(2, 1, 2)  # 2 rows, 1 column, 2nd subplot
plt.hist(female_population_percentage, bins=plot_config["hist_bins"], color=plot_config["hist_color"], edgecolor=plot_config["hist_edgecolor"])
plt.title('Female Population Ages 20-24 in India (Histogram)')
plt.xlabel('Female Population (%)')
plt.ylabel('Frequency')

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plots
plt.show()


```
